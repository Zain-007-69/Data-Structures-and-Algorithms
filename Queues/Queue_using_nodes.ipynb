# Queue Using Linked Nodes:

```python
class Node:
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next_node = next_node


class QueueUsingNodes:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, item):
        new_node = Node(item)
        if self.rear is None:
            self.front = self.rear = new_node
        else:
            self.rear.next_node = new_node
            self.rear = new_node

    def dequeue(self):
        if not self.is_empty():
            item = self.front.data
            self.front = self.front.next_node
            if self.front is None:
                self.rear = None
            return item
        return None

    def peek(self):
        if not self.is_empty():
            return self.front.data
        return None

    def size(self):
        count = 0
        current_node = self.front
        while current_node:
            count += 1
            current_node = current_node.next_node
        return count

    def display(self):
        current_node = self.front
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next_node
        print("None")

    def clear(self):
        self.front = self.rear = None

    def contains(self, item):
        current_node = self.front
        while current_node:
            if current_node.data == item:
                return True
            current_node = current_node.next_node
        return False

    def copy(self):
        new_queue = QueueUsingNodes()
        current_node = self.front
        while current_node:
            new_queue.enqueue(current_node.data)
            current_node = current_node.next_node
        return new_queue

    def reverse(self):
        prev_node = None
        current_node = self.front
        while current_node:
            next_node = current_node.next_node
            current_node.next_node = prev_node
            prev_node = current_node
            current_node = next_node
        self.front = prev_node


# Testing QueueUsingNodes
queue_nodes = QueueUsingNodes()

# Enqueue elements
queue_nodes.enqueue(1)
queue_nodes.enqueue(2)
queue_nodes.enqueue(3)

# Display queue
queue_nodes.display()

# Check if empty
print("Is Empty:", queue_nodes.is_empty())

# Peek at the front element
print("Peek:", queue_nodes.peek())

# Dequeue elements
print("Dequeue:", queue_nodes.dequeue())
print("Dequeue:", queue_nodes.dequeue())

# Display queue after dequeues
queue_nodes.display()

# Check size
print("Size:", queue_nodes.size())

# Check if it contains an element
print("Contains 2:", queue_nodes.contains(2))

# Clear the queue
queue_nodes.clear()
queue_nodes.display()
```
