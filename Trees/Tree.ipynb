# Tree

```python
class Tree:
    class Node:
        def __init__(self,data):
            self.data=data
            self.children=[]
    
    def __init__(self):
        self.root=None
        self.size=0
    
    def addnode(self,data,parent=None):
        if self.root==None and parent==None:
            self.root = self.Node(data)
            
        elif self.root==None and parent!=None:
            raise Exception('Add root firstly')
        
        elif self.root!=None and parent==None:
            raise Exception('Root is already exist')
        
        elif self.root!=None and parent!=None:
            if self.root.data==parent:
                self.root.children.append(self.Node(data))
        
            else:
                self.addnodeAux(self.root,data,parent)
        
        self.size+=1
    def addnodeAux(self,main,data,parent):
        for i in main.children:
            if i.data==parent:
                i.children.append(self.Node(data))
                return
            self.addnodeAux(i,data,parent)
    
    def print(self):
        if self.size!=0:
            if self.root!=None:
                print(self.root.data)
                
            elif self.root!=None:
                raise Exception('Tree does not exist')
            
            self.printAux(self.root)
    def printAux(self,main):
        if len(main.children)!=0:
            for i in main.children:
                print(i.data)
                self.printAux(i)
    
    def getsize(self):
        return self.size
    
    def leaves(self):
        if self.root==None:
            return 0
        if len(self.root.children)==0:
            return 1
        else:
            return  self.leavesAux(self.root)
    def leavesAux(self,main,count=0):
        if  len(main.children)==0:
            return count+1
        for i in main.children:
            count += self.leavesAux(i,count+1)
            return count
    
    def update(self,old,new):
        if self.root.data==old:
            self.root.data=new
        else:
            self.updateAux(self.root,old,new)
    def updateAux(self,main,old,new):
        for i in main.children:
            if i.data==old:
                i.data=new
            self.updateAux(i,old,new)
    
    def remove(self,val):
        if self.root.data==val:
            self.root=None
            self.size=0
            return
        self.removeAux(self.root,val)
    def removeAux(self,main,val):
        for i in range(len(main.children)-1):
            if main.children[i].data==val:
                main.children[i].children=[]
                del main.children[i]
            self.removeAux(main.children[i],val)
    
    def depth(self,val):
        if self.root.data==val:
            return 1
        return self.depthAux(self.root,val)
    def depthAux(self,main,val,count=2):
        for i in main.children:
            if i.data==val:
                return count
            return self.depthAux(i,val,count+1)
    
    def levelwiseprint(self):
        level=0
        print(' '*level,self.root.data)
        self.levelwiseprintAux(self.root,level+1)
    def levelwiseprintAux(self,node,level):
        if len(node.children)!=0:
            for i in node.children:
                print('   '*level,i.data)
                self.levelwiseprintAux(i,level+1)

def main():
    t=Tree()
    t.addnode(20)
    # add the element as child node to an existing node of tree
    t.addnode(5, 20)
    t.addnode(10, 20)
    t.addnode(15, 20)
    t.addnode(7, 5)
    t.addnode(8, 5)
    t.addnode(9, 10)
    t.addnode(11, 15)
    t.addnode(12, 15)
    t.addnode(13, 15)
    t.addnode(14, 7)
    t.addnode(15, 7)
    t.addnode(16, 11)
    t.levelwiseprint()
main()
```
