# Singly Linked List using Linked Nodes:

```python
class Node:
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next_node = next_node


class SinglyLinkedListUsingNodes:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next_node:
                current_node = current_node.next_node
            current_node.next_node = new_node

    def display(self):
        current_node = self.head
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next_node
        print("None")

    def length(self):
        count = 0
        current_node = self.head
        while current_node:
            count += 1
            current_node = current_node.next_node
        return count

    def is_empty(self):
        return self.head is None

    def get_element_at_index(self, index):
        count = 0
        current_node = self.head
        while current_node and count < index:
            count += 1
            current_node = current_node.next_node
        if current_node:
            return current_node.data
        return None

    def contains(self, data):
        current_node = self.head
        while current_node:
            if current_node.data == data:
                return True
            current_node = current_node.next_node
        return False

    def remove_at_index(self, index):
        if index == 0:
            self.head = self.head.next_node
        else:
            count = 0
            current_node = self.head
            while current_node and count < index - 1:
                count += 1
                current_node = current_node.next_node
            if current_node and current_node.next_node:
                current_node.next_node = current_node.next_node.next_node

    def remove_element(self, data):
        current_node = self.head
        if current_node and current_node.data == data:
            self.head = current_node.next_node
        else:
            while current_node and current_node.next_node and current_node.next_node.data != data:
                current_node = current_node.next_node
            if current_node and current_node.next_node:
                current_node.next_node = current_node.next_node.next_node

    def clear(self):
        self.head = None

    def insert_at_index(self, index, data):
        new_node = Node(data)
        if index == 0:
            new_node.next_node = self.head
            self.head = new_node
        else:
            count = 0
            current_node = self.head
            while current_node and count < index - 1:
                count += 1
                current_node = current_node.next_node
            if current_node:
                new_node.next_node = current_node.next_node
                current_node.next_node = new_node


# Testing SinglyLinkedListUsingNodes
sll_nodes = SinglyLinkedListUsingNodes()

# Append elements
sll_nodes.append(1)
sll_nodes.append(2)
sll_nodes.append(3)

# Display linked list
sll_nodes.display()

# Check length
print("Length:", sll_nodes.length())

# Check if empty
print("Is Empty:", sll_nodes.is_empty())

# Get element at index
print("Element at index 1:", sll_nodes.get_element_at_index(1))

# Check if it contains an element
print("Contains 2:", sll_nodes.contains(2))

# Remove element at index
sll_nodes.remove_at_index(1)
sll_nodes.display()

# Remove specific element
sll_nodes.remove_element(3)
sll_nodes.display()

# Clear the linked list
sll_nodes.clear()
sll_nodes.display()
```
