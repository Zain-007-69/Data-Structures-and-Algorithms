# Circluar Doubly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyCircularLinkedList:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head is None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            temp = self.head.prev
            temp.next = new_node
            new_node.prev = temp
            new_node.next = self.head
            self.head.prev = new_node

    def prepend(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            temp = self.head.prev
            temp.next = new_node
            new_node.prev = temp
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def delete(self, key):
        if not self.head:
            return

        if self.head.data == key:
            if self.head.next == self.head:
                self.head = None
            else:
                temp = self.head.prev
                temp.next = self.head.next
                self.head.next.prev = temp
                self.head = self.head.next
        else:
            current_node = self.head.next
            while current_node != self.head:
                if current_node.data == key:
                    current_node.prev.next = current_node.next
                    current_node.next.prev = current_node.prev
                    current_node = None
                    return
                current_node = current_node.next

    def display(self):
        if not self.head:
            return

        current_node = self.head
        while True:
            print(current_node.data, end=" <-> ")
            current_node = current_node.next
            if current_node == self.head:
                break
        print("Head")

# Example usage:
doubly_circular_linked_list = DoublyCircularLinkedList()
doubly_circular_linked_list.append(1)
doubly_circular_linked_list.append(2)
doubly_circular_linked_list.append(3)
doubly_circular_linked_list.prepend(0)
doubly_circular_linked_list.display()

doubly_circular_linked_list.delete(2)
doubly_circular_linked_list.display()
```
