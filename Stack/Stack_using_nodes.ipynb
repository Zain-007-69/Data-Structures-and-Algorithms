# Stack using Linked Nodes:

```python
class Node:
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next_node = next_node


class StackUsingNodes:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top is None

    def push(self, item):
        new_node = Node(item, self.top)
        self.top = new_node

    def pop(self):
        if not self.is_empty():
            item = self.top.data
            self.top = self.top.next_node
            return item
        return None

    def peek(self):
        if not self.is_empty():
            return self.top.data
        return None

    def size(self):
        count = 0
        current_node = self.top
        while current_node:
            count += 1
            current_node = current_node.next_node
        return count

    def display(self):
        current_node = self.top
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next_node
        print("None")

    def clear(self):
        self.top = None

    def contains(self, item):
        current_node = self.top
        while current_node:
            if current_node.data == item:
                return True
            current_node = current_node.next_node
        return False

    def copy(self):
        new_stack = StackUsingNodes()
        current_node = self.top
        while current_node:
            new_stack.push(current_node.data)
            current_node = current_node.next_node
        return new_stack

    def reverse(self):
        prev_node = None
        current_node = self.top
        while current_node:
            next_node = current_node.next_node
            current_node.next_node = prev_node
            prev_node = current_node
            current_node = next_node
        self.top = prev_node
```

```python
# Testing StackUsingNodes
stack_nodes = StackUsingNodes()

# Push elements
stack_nodes.push(1)
stack_nodes.push(2)
stack_nodes.push(3)

# Display stack
stack_nodes.display()

# Check if empty
print("Is Empty:", stack_nodes.is_empty())

# Peek at the top element
print("Peek:", stack_nodes.peek())

# Pop elements
print("Pop:", stack_nodes.pop())
print("Pop:", stack_nodes.pop())

# Display stack after pops
stack_nodes.display()

# Check size
print("Size:", stack_nodes.size())

# Check if it contains an element
print("Contains 2:", stack_nodes.contains(2))

# Clear the stack
stack_nodes.clear()
stack_nodes.display()
```
